### O Problema

#### **Problema a Ser Resolvido**
A escassez de profissionais de saúde mental nos Estados Unidos, impactando o acesso da população a serviços essenciais. Regiões com deficit de profissionais podem enfrentar maiores taxas de transtornos mentais não tratados, aumentando a sobrecarga em emergências hospitalares e reduzindo a qualidade de vida da população. Este estudo busca analisar padrões de escassez de profissionais de saúde mental com base nos dados de Áreas de Escassez de Profissionais de Saúde (HPSA), fornecidos pela Health Resources and Services Administration (HRSA). O objetivo é identificar tendências, correlacionar fatores socioeconômicos e avaliar impactos para subsidiar políticas públicas e estratégias de alocação de recursos.

#### **Fonte de Dados**
A base de dados utilizada provém da HRSA, disponível em (https://data.hrsa.gov/data/download), contendo informações sobre áreas geográficas, populações e instalações com escassez de serviços de saúde. Os principais conjuntos de dados incluem:

- **All HPSAs – CSV.csv**: Contém HPSAs relacionados à saúde mental.
- **MUA/P – CSV.csv**: Lista áreas/populações medicamente carentes.

Este estudo visa fornecer insights baseados em dados para apoiar decisões estratégicas e melhorar a distribuição de recursos na saúde mental.

## **Resumo**

Este estudo analisa a escassez de profissionais de saúde mental nos Estados Unidos e seu impacto na oferta de serviços essenciais. Utilizando dados da HRSA, o objetivo foi desenvolver um modelo preditivo para estimar a pontuação HPSA com base em variáveis demográficas e de atendimento. Indicadores socioeconômicos foram inicialmente considerados, mas excluídos devido à alta taxa de valores ausentes. Foram avaliados três modelos de classificação – Random Forest, Decision Tree e KNN – otimizados com GridSearchCV e validação cruzada. O Random Forest obteve o melhor desempenho, com acurácia de 84,52%, destacando-se pela capacidade de identificar áreas com maior necessidade de recursos. Os achados reforçam a importância da modelagem preditiva para subsidiar políticas públicas e otimizar a distribuição de profissionais e infraestrutura de saúde.

#### **6. Análise de colunas com alta taxa de valores ausentes (> 40% de valores ausentes)**  

Será verificado se há colunas com mais de **40% de valores ausentes**. Será realizada uma avaliação para determinar se é viável preenchê-las utilizando estratégias como:  

- **Média/mediana/moda**: Para colunas numéricas, verificaremos se a imputação com média ou mediana mantém a integridade dos dados.  
- **Preenchimento baseado em categorias**: Para colunas categóricas, avaliaremos se a substituição pela moda ou por valores inferidos a partir de outras colunas faz sentido.  
- **Relevância da informação**: Mesmo com valores ausentes, algumas colunas podem conter informações críticas para o modelo, justificando sua manutenção com técnicas adequadas de imputação.  

Após essa análise, caso não seja possível um preenchimento adequado sem introduzir viés, as colunas poderão ser removidas.

```python
# Verificar percentual de valores nulos
missing_percent = hpsa_data.isnull().sum() / len(hpsa_data) * 100

# Colunas com mais de 40% de valores nulos
high_missing = missing_percent[missing_percent > 40]
```

```python
print("\nColunas com mais de 40% de valores nulos:")

plt.figure(figsize=(12, 6))
sns.barplot(x=high_missing.index, y=high_missing.values, hue=high_missing.values, legend=False)
plt.xticks(rotation=90)
plt.xlabel("Colunas do Conjunto de Dados")
plt.ylabel("Quantidade de Valores Ausentes")
plt.title("Quantidade de Dados Ausentes por Coluna")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Criar um dicionário para sugerir estratégias de imputação
imputation_strategies = {}

for col in high_missing.index:
    if hpsa_data[col].dtype in ['int64', 'float64']:
        # Para colunas numéricas, verificar se a média ou mediana são viáveis
        imputation_strategies[col] = {
            'mean': hpsa_data[col].mean(),
            'median': hpsa_data[col].median()
        }
    else:
        # Para colunas categóricas, verificar a moda
        imputation_strategies[col] = {
            'mode': hpsa_data[col].mode()[0] if not hpsa_data[col].mode().empty else None
        }

imputation_df = pd.DataFrame(imputation_strategies).T
imputation_df.index.name = "Column"
imputation_df
```

#### Descidindo entre preencher com **Média** ou **mediana**.

Usamos a função `skew` da biblioteca `scipy.stats` para calcular a assimetria dos dados. A função retorna o coeficiente de assimetria, calculado atráves do **coeficiente de assimetria de *Fisher-Pearson***, que nos informa se os dados estão inclinados para a esquerda ou direita.

Decisão:

*   Se a **assimetria é baixa** (menor que 1 em termos absolutos), indicamos que a distribuição é mais simétrica, então a **média** é uma boa escolha para preencher os valores nulos.
*   Se a **assimetria é alta** (maior que 1 em termos absolutos), indicamos que a distribuição é mais distorcida devido a outliers, e a **mediana** é uma escolha mais robusta.


```python
# Definindo as colunas numéricas que precisam ser preenchidas
colunas_numericas = [
    "HPSA FTE",
    "% of Population Below 100% Poverty",
    "HPSA Estimated Served Population",
    "HPSA Estimated Underserved Population",
    "HPSA Shortage",
    "HPSA Degree of Shortage",
    "HPSA Provider Ratio Goal",
    "HPSA Designation Population"
]
```

```python
from scipy.stats import skew, kurtosis

# função para decidir entre média e mediana
def decide_metric(df, col):
    data = df[col].dropna()

    # calcular a média e a mediana
    mean = data.mean()
    median = data.median()

    # calcular a assimetria e a curtose
    skewness = skew(data)
    kurt = kurtosis(data)

    # Critério de escolha baseado na assimetria
    if abs(skewness) > 1:
        decision = 'Mediana (distribuição assimétrica)'
        fill_value = df[col].median()
    else:
        decision = 'Média (distribuição simétrica ou normal)'
        fill_value = df[col].mean()

    # preenche os valores ausentes com o valor calculado
    df[col] = df[col].fillna(fill_value)


    print(f"Coluna: {col}")
    print(f"  - Média: {mean:.2f}")
    print(f"  - Mediana: {median:.2f}")
    print(f"  - Assimetria: {skewness:.2f}")
    print(f"  - Curtose: {kurt:.2f}")
    print(f"  - Decisão: {decision}")

    print(f"\nColuna: {col} - Preenchendo com: {decision} (Valor: {fill_value:.2f})")
    print("-" * 50)

    return decision

# Aplicando a decisão para cada coluna
for coluna in colunas_numericas:
    decide_metric(hpsa_data, coluna)
```

### **One-Hot Encoding**

O **One-Hot Encoding** é uma técnica para transformar variáveis categóricas em formato binário (0 ou 1), necessária para algoritmos de machine learning, que não podem trabalhar com dados textuais diretamente.

### Justificativas:
1. **Variáveis Categóricas Não Ordenadas**: As variáveis como `'HPSA Status'` e `'Metropolitan Indicator'` não têm uma ordem intrínseca, e o One-Hot Encoding evita a interpretação errada de hierarquia.
2. **Compatibilidade com Modelos de Machine Learning**: A maioria dos modelos não pode trabalhar com variáveis categóricas, então elas precisam ser transformadas para um formato numérico.
3. **Evita Multicolinearidade**: Usando `drop_first=True`, evita a criação de uma coluna redundante e possíveis problemas de multicolinearidade.

```python
for col in hpsa_data_cleaned.select_dtypes(include=['object']) :
    print(col, ':', len(hpsa_data_cleaned[col].unique()), 'labels' )
```

```python
# Aplicando o One-Hot Encoding nas colunas categóricas
categorical_columns = hpsa_data_cleaned.select_dtypes(include=['object']).columns

# Criando as variáveis One-Hot Encoded
hpsa_data_encoded = pd.get_dummies(hpsa_data_cleaned, columns=categorical_columns, drop_first=True)

# Exibindo as primeiras linhas do novo DataFrame
hpsa_data_encoded.head()
```

### **Padronização de Dados**

Para realizar a padronização (ou **normalização**) dos dados, vamos aplicar um procedimento, utilizando o `StandardScaler`, que visa transformar as variáveis numéricas para que tenham uma média de 0 e um desvio padrão de 1, de modo a evitar que variáveis com escalas maiores dominem a análise.

#### Justificativa para a padronização:
A padronização dos dados é importante por várias razões:

1. **Evitar influência de variáveis em escalas diferentes**: Se algumas variáveis possuem valores muito maiores que outras (por exemplo, uma variável que varia de 0 a 1000 e outra que varia de 0 a 1), os modelos podem dar mais peso às variáveis com maior amplitude, o que pode prejudicar a análise e a performance do modelo.

2. **Melhor desempenho em modelos de aprendizado de máquina**: Modelos como regressão logística, KNN (K-Nearest Neighbors) e SVM (Support Vector Machines) frequentemente se beneficiam da padronização, pois os coeficientes dos modelos são mais estáveis e o treinamento converge mais rápido.

3. **Interpretação consistente**: Quando as variáveis estão na mesma escala, fica mais fácil comparar e interpretar a relação entre elas durante o processo de modelagem.

```python
from sklearn.preprocessing import StandardScaler

# Inicializando o escalador
scaler = StandardScaler()

# Aplicando a padronização
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

# aplicando a transformação scaler ao conjunto de teste
X_test_scaled = scaler.transform(X_test)
```

## **Etapa inicial de Classificação**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Lista de modelos para testar
modelos = {
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier()
}

resultados = []

for nome, modelo in modelos.items():
    modelo.fit(X_train_scaled, y_train)
    y_pred = modelo.predict(X_test_scaled)

    # Calcular métricas
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted', zero_division=1)
    rec = recall_score(y_test, y_pred, average='weighted', zero_division=1)
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Armazenar os resultados
    resultados.append({"Modelo": nome, "Acurácia": acc, "Precisão": prec, "Recall": rec, "F1-Score": f1})

df_resultados = pd.DataFrame(resultados)

# Ordenar pelo melhor F1-Score
df_resultados = df_resultados.sort_values(by="F1-Score", ascending=False)

df_resultados
```

Podemos adicionar **GridSearchCV** e **StratifiedKFold** para otimizar os hiperparâmetros dos modelos e garantir uma avaliação mais robusta. O **GridSearchCV** permite testar várias combinações de hiperparâmetros para encontrar a melhor configuração, enquanto o **StratifiedKFold** divide os dados em k partes para validação cruzada, reduzindo o risco de overfitting.

### **Hiperparâmetros Testados**

1. **Random Forest:**
   - **`n_estimators`**: Número de árvores na floresta. Testamos valores como 50, 100 e 200 para encontrar um equilíbrio entre desempenho e tempo de treinamento.
   - **`max_depth`**: Profundidade máxima das árvores. Valores como `None`, 10 e 20 ajudam a controlar o overfitting.
   - **`min_samples_split`**: Número mínimo de amostras necessárias para dividir um nó. Testamos 2, 5 e 10 para evitar divisões muito específicas.

3. **Decision Tree:**
   - **`max_depth`**: Profundidade máxima da árvore. Valores como `None`, 10 e 20 ajudam a evitar overfitting.
   - **`min_samples_split`**: Número mínimo de amostras para dividir um nó. Testamos 2, 5 e 10.
   - **`criterion`**: Critério de divisão. "gini" e "entropy" são as métricas mais comuns para árvores de decisão.

4. **KNN:**
   - **`n_neighbors`**: Número de vizinhos. Testamos 3, 5 e 7 para encontrar o valor ideal.
   - **`weights`**: Peso dos vizinhos. "uniform" trata todos os vizinhos igualmente, enquanto "distance" dá mais peso aos vizinhos mais próximos.
   - **`metric`**: Métrica de distância. "euclidean" e "manhattan" são as mais utilizadas.

---

### **Vantagens do Uso de GridSearchCV e KFold**
1. **Otimização de Hiperparâmetros:**
   - O GridSearchCV testa todas as combinações de hiperparâmetros para encontrar a melhor configuração.
   - Isso melhora o desempenho do modelo e evita escolhas subótimas.

2. **Validação Cruzada:**
   - O StratifiedKFold divide os dados em k partes, garantindo que o modelo seja avaliado em diferentes subconjuntos dos dados.
   - Isso reduz o risco de overfitting e fornece uma avaliação mais confiável.

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

modelos = {
    "Random Forest": {
        "modelo": RandomForestClassifier(),
        "parametros": {
            "n_estimators": [50, 100, 200],  # Número de árvores na floresta
            "max_depth": [None, 10, 20],     # Profundidade máxima das árvores
            "min_samples_split": [2, 5, 10]  # Número mínimo de amostras para dividir um nó
        }
    },
    "Decision Tree": {
        "modelo": DecisionTreeClassifier(),
        "parametros": {
            "max_depth": [None, 10, 20],     # Profundidade máxima da árvore
            "min_samples_split": [2, 5, 10], # Número mínimo de amostras para dividir um nó
            "criterion": ["gini", "entropy"] # Critério de divisão
        }
    },
    "KNN": {
        "modelo": KNeighborsClassifier(),
        "parametros": {
            "n_neighbors": [3, 5, 7],        # Número de vizinhos
            "weights": ["uniform", "distance"],  # Peso dos vizinhos
            "metric": ["euclidean", "manhattan"]  # Métrica de distância
        }
    }
}
best_models = {}
resultados = []

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for nome, config in modelos.items():
    print(f"Treinando {nome}...")

    # GridSearchCV para otimização de hiperparâmetros
    grid_search = GridSearchCV(
        estimator=config["modelo"],
        param_grid=config["parametros"],
        scoring="f1_weighted",
        cv=kfold,
        n_jobs=-1
    )

    # Treinar o modelo com GridSearchCV
    grid_search.fit(X_train_scaled, y_train)
    # Melhores hiperparâmetros encontrados
    print(f"Melhores hiperparâmetros para {nome}: {grid_search.best_params_}")

    y_pred = grid_search.predict(X_test_scaled)

    # Salva o melhor modelo
    best_models[nome] = grid_search.best_estimator_

    # métricas
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted', zero_division=1)
    rec = recall_score(y_test, y_pred, average='weighted', zero_division=1)
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Armazenar os resultados
    resultados.append({
        "Modelo": nome,
        "Acurácia": acc,
        "Precisão": prec,
        "Recall": rec,
        "F1-Score": f1,
        "Melhores Hiperparâmetros": grid_search.best_params_
    })

df_resultados = pd.DataFrame(resultados)

# Ordenar pelo melhor F1-Score
df_resultados = df_resultados.sort_values(by="F1-Score", ascending=False)

df_resultados
```
